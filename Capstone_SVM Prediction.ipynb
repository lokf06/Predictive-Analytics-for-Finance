### Step 3: Preparing the data for training and testing

1) Divide independent and dependent variable into two separate variables.

2) Split the data into training and testing datasets.

```python
# Define X and y as the predictors (independent X) and target variables (dependent y) where:

#X contains all other columns except for default_coded
#y contains only default_coded column

X = df2.drop(['default_coded'], axis=1)   # input (independent variable)
y = df['default_coded']                  # output (dependent variable)
```

```python
# 2) Splitting our data into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X ,y, 
                                                    test_size = 0.2, 
                                                    shuffle=True, 
                                                    stratify=y,
                                                    random_state=SEED)
```

### Step 4: Training the model

```python
# Import the Support Vector Machines algorithm

from sklearn.svm import LinearSVC
SVM_model = LinearSVC()
```

```python
# Hyperparameter tuning using K-fold cross validation
# ... via Grid Search method

from sklearn.model_selection import GridSearchCV

param_grid = {'C': np.logspace(-4.5, -1.3, 10),  # C = 0.00003162 to 0.0501
              'penalty': ['l1', 'l2']}


gs_SVM_model = GridSearchCV(SVM_model,
                      param_grid, 
                      cv=5,
                      scoring='f1_macro',
                      n_jobs=-1)
gs_SVM_model.fit(X_train, y_train)
print("'{}' gives the best F1-score at: {:.2%}".format(gs_SVM_model.best_params_, gs_SVM_model.best_score_))
```

```python
# Training the final model using the best parameters from above fine-tuning

SVM_model2 = classifier = LinearSVC(C=0.0000717, penalty='l2')
SVM_model2.fit(X_train, y_train)
```

### Step 5: Testing the model

```python
# testing the X_test dataset

y_pred = classifier.predict(X_test)

compare_df2 = pd.DataFrame({"Desired Output (Actuals)": y_test, 
                            "Predicted Output": y_pred})

# add a 'Match' column
compare_df2['Match'] = compare_df2['Desired Output (Actuals)'] == compare_df2['Predicted Output']

#print(compare_df2)

compare_df2
```

### Step 6: Evaluate the model

```python
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score

# Classification Report
print("Classification report:")
#print()

#print(classification_report(y_test, y_pred, target_names=class_names.tolist()))
#print()

print(classification_report(y_test, y_pred))

# Confusion Matrix
print("Confusion Matrix:")
confusion_matrix(y_test, y_pred)
```

```python
# # list out the result of Accuracy, Precision and Recall

test_acc = classifier.score(X_test, y_test)
print("Accuracy  = {:.2f}".format(test_acc))

# Precision
test_pre = precision_score(y_test, y_pred, average=None)
print("Precision = {}".format(test_pre))

# Recall
test_rec  = recall_score(y_test, y_pred, average=None)
print("Recall    = {}".format(test_rec))
```

```python
# Plot the confusion matrix using Seaborn library

plt.figure(figsize=(7,5))
_ = sns.heatmap(confusion_matrix(y_test, y_pred), 
                annot=True, annot_kws={"size": 15},
                cmap=plt.cm.Blues,fmt='.0f')
                #xticklabels=class_names, yticklabels=class_names)
_ = plt.ylabel('True Label', fontweight='bold')
_ = plt.xlabel('Predicted Label', fontweight='bold')
```

```python
#check for overfitting on both training and testing sets

from sklearn.metrics import accuracy_score

# Make predictions on the training and testing sets
y_train_pred = SVM_model2.predict(X_train)
y_test_pred = SVM_model2.predict(X_test)

# Calculate accuracy scores
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)

# Check for overfitting
if train_accuracy > test_accuracy:
    print("Model may be overfitting.")
else:
    print("Model generalizes well.")
```
